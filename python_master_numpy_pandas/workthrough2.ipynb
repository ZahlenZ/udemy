pro tip.
numpys select() function lets you create columns based on multiple conditions

conditions = (
    (logical) & (logical),
    (logical) | (logical)
)

choices = ["category1", "category2"]

df["new_column"] = np.select(conditions, choices, default = "No")

will evaluate the value of the new column based on the conditionals and assign the choices

```python
df.head()
```

```python
conditions = (
    df["date"].str[5: 7] == "12",
    ((df["date"].str[5: 7] == "05") & (df["date"].str[-2:] == "07")),
    ((df["date"].str[5: 7] == "07") & (df["date"].str[-2:] == "01"))
)
choices = ["Holiday Bonus", "Corporate Month", "Summer Special"]

df["seasonal_bonus"] = np.select(conditions, choices, default = "None")
```

```python
df["seasonal_bonus"].value_counts().iloc[1:].sum() * 100
```

```python
df.drop(
    ["pct_to_target", "met_target", "bonus_payable", "seasonal_bonus"], 
    axis = 1, 
    inplace = True
)
```

```python
conditions = (
    df["date"].str[5: 7] == "12",
    ((df["date"].str[5: 7] == "05") & (df["date"].str[-2:] == "07")),
    ((df["date"].str[5: 7] == "07") & (df["date"].str[-2:] == "01"))
)
choices = ["Holiday Bonus", "Corporate Month", "Summer Special"]

df["seasonal_bonus"] = np.select(conditions, choices, default = "None")

df["pct_to_target"] = df["transactions"] / 2500
df["met_target"] = df["pct_to_target"] >= 1
df["bonus_payable"] = df["met_target"] * 100
```

```python
df.head()
```

```python
df = df.assign(
    target_pct = df["transactions"] / 2500,
    met_target = (df["transactions"] / 2500) >= 1,
    bonus_payable = ((df["transactions"] / 2500) >= 1) * 100,
    seasonal_bonus = np.select(conditions, choices, default = "None")
)
```
