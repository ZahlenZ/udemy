time series data allow for method for fixing missing data beyond .fillna()

dates.ffill() - grab last value and fill forward
dates.bfill() - grab next value and fill backward
date.interpolate() - must first convert the dates to type float, and then it will do a linear estimation of the missing value

you can aggregate time series using the .groupby() method, but depending on use case there are better methods.
sales.groupby(df["date"].dt.month).agg({"sales": "sum"}) goes over all months that are the same(which means it grabs month 1 for all years)

resampling is a special form of time series aggregation that modifies date indices and fills any faps to create a continuous date values

dates.resample("M").sum()

dates.resmaple("M").transform("sum") - use transform() to keep the original index and apply the group-level aggregations
