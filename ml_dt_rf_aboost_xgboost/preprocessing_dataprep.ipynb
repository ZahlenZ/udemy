```python
upper_limit = np.percentile(df["n_hot_rooms"], [99])[0]
```

```python
df.loc[df["n_hot_rooms"] > 2 * upper_limit]
```

```python
df.loc[df["n_hot_rooms"] > 3 * upper_limit, "n_hot_rooms"] = 3 * upper_limit
```

```python
lower_limit = np.percentile(df["rainfall"], [1])[0]
```

```python
lower_limit * .3
```

```python
df.loc[df["rainfall"] < .3 * lower_limit, "rainfall"] = 0.3 * lower_limit
```
