```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../data/HotelCustomersDataset.xlsx")
```

```python
df.head()
```

```python
sns.histplot(
    x="LodgingRevenue",
    data=df.query("LodgingRevenue < 1500")
)
```

```python
top5 = df.groupby("Nationality")["LodgingRevenue"].sum().sort_values(ascending=False).iloc[0:5]
```

```python
sns.barplot(
    x="RoomNights",
    y="Nationality",
    data=df[df["Nationality"].isin(top5.index)],
    errorbar=None,
    hue="Nationality",
    palette="husl",
    estimator=np.mean
)

sns.despine()
```

```python
df.head()
```

```python
sns.boxplot(
    y="LodgingRevenue",
    data=df.query("LodgingRevenue < 1500")
)
```

```python
sns.violinplot(
    x="Nationality",
    y="Age",
    data=df[df["Nationality"].isin(top5.index)],
    hue="Nationality",
    palette="viridis"
)
```

```python
sns.violinplot(
    x="Nationality",
    y="LodgingRevenue",
    data=df[df["Nationality"].isin(top5.index)].query("LodgingRevenue < 1500")
)
```

```python
df.head()
```

```python
sns.regplot(
    x="RoomNights",
    y="LodgingRevenue",
    data=df
)

plt.show()
```

```python
sns.lmplot(
    data=df[df["Nationality"].isin(top5.index)].query("LodgingRevenue < 1500 and RoomNights < 15"),
    x="RoomNights",
    y="LodgingRevenue",
    hue="Nationality"
)
plt.show()
```
